# Analogy between electric and mechanical oscillators

Motivated by the mathematical analogy between acoustic and electromagnetic waves, some researchers have attempted to find the counterpart left-handed acoustic metamaterials. Basically, it is to find metamaterials that possess negative mass density and negative modulus [4â8]. However, in contrast to the natural negative electric permittivity, no natural material has been found to possess negative mass density or negative modulus. Ever since the investigation on locally resonant acoustic materials [9], some new developments have further been theoretically attempted and shown that metamaterials with certain manmade microstructures may exhibit negative mass/mass density if treated as a classical elastic solid.



# Discrete mass-spring system



The mechanism of negative effective mass can be discovered by the mass-spring structure shown in Fig. 1, where a hollow sphere of the weight $M_0$ connects a rigid sphere of weight m by two massless and elastic springs with equal spring cofficient G. In the time harmonic case, the equilibrium equation of the inner mass m results in

$$
\dfrac{x}{X}=\dfrac{\omega_1^2}{\omega_1^2-\omega^2}
$$

where x and X are respectively the displacement of the 
inner and outer mass, and $Ï_1 = \sqrt{2G/m}$ is the resonant frequency. The composite hollow structure can be considered as an equivalent solid object and the effective mass $M_{eff}$ of the unit can be defined as the total momentum divided by the velocity of the outer mass

$$
M_{eff}=M_0+\dfrac{m \omega_1^2}{\omega_1^2-\omega^2}
$$

Due to the negative momentum of the inner mass
overwhelming the system, the effective mass $M_{eff}$  takes
negative values at frequencies ranging from $Ï_0$ to $Ï_0 \sqrt{(M_0 + m)/M_0}$. When the inner mass is fixed, effective dynamic mass  becomes

$$
M_{eff}=M_0 \left(1-\dfrac{\omega_0^2}{\omega^2} \right)
$$

where $Ï_0 = \sqrt{2G/M_0}$.

## Negative effective mass in lattices


The negative effective mass/mass density is the result of  modeling of acoustic metamaterials. In the present study, we consider a one-dimensional lattice consisting of lattice masses each of which contains an internal mass. This simple lattice system is employed to illustrate how the negative effective mass could occur and what effects it has on wave propagation in the lattice system. 

A one-dimensional model can be used to introduce the negative effective mass effect as shown in Figure below. If it is to be represented by a single mass, then the effective mass must be defined in order that the effective mass $m_{eff}$ would produce the same motion of $m_1$. In other words, the identity of the internal mass $m_2$ would be ignored and its effect would be absorbed by the introduction of an effective mass $m_{eff}$.



The lattice system is constructed by connecting the unit cells of Figure above with linear springs. The unit-cells are placed periodically at a spacing of L. Harmonic wave propagation in the mass-in-mass lattice system is considered. The equations of motion for the j-th unit cell are

$$
m_1^{(j)} \dfrac{d^2 u_1^{(j)} }{dt^2}+k_1 (2 u_1^{(j)}-u_1^{(j-1)}-u_1^{(j+1)}) +k_2 (u_1^{(j)}-u_2^{(j)})=0
$$

$$
m_2^{(j)} \dfrac{d^2 u_2^{(j)} }{dt^2}+k_2 (u_2^{(j)}-u_1^{(j)})=0
$$
