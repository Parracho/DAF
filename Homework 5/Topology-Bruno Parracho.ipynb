__EXERCISE 1:__

Find the energy spectrum for the small SSH chain defined above as a function of $t_1/t_2$ (where $t_1=v$ and $t_2=w$).

The hamiltonian of a SSH chain is described by the following matrix, 
$$\hat{H}=\left(\begin{array}{ccc} 0 & t_1 & 0 \\ t_1 & 0 & t_2 \\ 0 & t_2 & 0\end{array}\right)$$
and it can be expanded to higher orders maintaining the pattern:

The energy spectrum are the eigenvalues of the hamiltonian, for each $v$.

```matlab
N=9;
v=0:0.1:2;
va=repmat(v,N+1,1);
length(v);
val=zeros(N+1,length(v));
n=1;
for u=v;
    H=zeros(N,N);
    for i=1:2:N 
        H(i,i+1)=u;
        H(i+1,i)=u;
    end
    for i=2:2:N 
        H(i,i+1)=1;
        H(i+1,i)=1;
    end
    val(:,n)=eig(H);
    n=n+1;
end
```

```matlab
plot(transpose(va),transpose(val))
grid on
title('E/w vs v/w')
xlabel('v/w') 
ylabel('E/w') 
```

### The winding vector

The bulk momentum-space Hamiltonian $\hat{H}(k)$ of any two-band model
(i.e., a model with 2 internal states per unit cell), reads

\begin{align}
H(k) &= d_x(k) \hat{\sigma}_x + d_y(k) \hat{\sigma}_y + d_z(k) \hat{\sigma}_z =
d_0(k) \hat{\sigma}_0 + d(k)\hat{\sigma}.
\end{align}

For the SSH model, $d_0(k)=0$, and the real numbers
$d_{x,y,z}\in\mathbb{R}$, the components of the $k$-dependent
3-dimensional vector $h(k)$, read

\begin{align}
\label{eq:ssh_bulk_h}
d_x(k) &= v + w \cos k;&
d_y(k) &= w \sin k;&
d_z(k) &= 0. 
\end{align}


__EXERCISE 2:__

Obtain all plots for $\hat{H}(k)$ vs. $k$ and $dy$ vs. $dx$. Where $\left\{\begin{array}{c} w=0 \\ v>w \\ v=w \\ v<w \\ v=0 \end{array}\right.$

```matlab
v=[1,1,1,0.6,0];
w=[0,0.6,1,1,1];
k=-pi:0.01:pi;
sigmax=[0,1;1,0];
sigmay=[0,-1*j;1*j,0];
val=zeros(2,length(k));
ki=repmat(k,2,1);
```

## First Case (w=0)

```matlab
for i=1:length(k)
    dx1(i)=v(1)+w(1)*cos(k(i));
    dy1(i)=w(1)*sin(k(i));
    H=dx1(i).*sigmax+dy1(i).*sigmay;
    val1(:,i)=eig(H);
end

plot(transpose(ki),transpose(val1),"-r")
grid on
title('E(k) vs k')
xlabel('k') 
ylabel('E(k)') 
legend("w=0")
xlim([-4 4])
ylim([-1.5 1.5])
```

```matlab
plot(dx1,dy1,"-o")
grid on
title('dy vs dx')
xlabel('dx') 
ylabel('dy') 
legend("w=0")
xlim([-2 2])
ylim([-2 2])
```

__EXERCISE 3:__

Plot the edge state amplitudes for several values of $v/w$.

The edge states are the eigenvectors of the hamiltonian for each $v$, we are only going the middle eigenvector.

```matlab
N=9; %N+1
v=0.1:0.1:0.5;
n=0;
vecs_total=zeros(N+1,(N+1)*length(v));
for u=v;
    H=zeros(N+1,N+1);
    for i=1:2:N 
        H(i,i+1)=u;
        H(i+1,i)=u;
    end
    for i=2:2:N
        H(i,i+1)=1;
        H(i+1,i)=1;
    end
    [vecs,D]=eig(H);
    vecs_total(1:N+1,n+1)=vecs(:,round(length(vecs))/2);
    n=n+1;
end
plot(vecs_total)
grid on
title('E/w vs v/w')
xlabel('v/w') 
ylabel('E/w') 
```
