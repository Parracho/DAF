We write the vector potential $$\vec{A}=A_x(x,y,z)\hat{x}+A_y(x,y,z)\hat{y}+A_z(x,y,z)\hat{z}$$ as follows

```
A = {A1[x, y, z], A2[x, y, z], A3[x, y, z]}
```

The free Energy expression, is$$ F =\alpha |\psi|^2 + \frac{\beta}{2} |\psi|^4 + \frac{1}{2m^*} \left| \left(-i\hbar\nabla - \frac{q}{c}\mathbf{A} \right) \psi \right|^2,$$ where the are no fields.

```
(* free energy *) (*Field zero*)
eq1 = \[Alpha] (\[Psi][x, y, z] \[Psi]c[x, y, z]) +
   1/2 \[Beta] (\[Psi][x, y, z]^2 \[Psi]c[x, y, z]^2) +
   1/(2 m) ((hbar/I Grad[\[Psi][x, y, z], {x, y, z}] - 
        q/c A \[Psi][x, y, z]).(-(hbar/I)
           Grad[\[Psi]c[x, y, z], {x, y, z}] - 
        q/c A \[Psi]c[x, y, z])) // Expand
```

To determine the minimization conditions, let's evaluate it's extremes by $$\frac{\partial F}{\partial \psi}=0 \textrm{ and } \frac{\partial F}{\partial \psi^*}=0$$

```
(*DErivada em relacao a psi*)
eq2 = VariationalD[eq1, \[Psi]c[x, y, z], {x, y, z}] // Expand
```

```
eq3 = VariationalD[eq1, \[Psi][x, y, z], {x, y, z}] // Expand
```

Considering now the free Energy expression with a non-zero field, $\vec{B}$, and also knowing that $\vec{B}=\vec{\nabla} \times \vec{A}$ we get $$ F = F_n + \alpha |\psi|^2 + \frac{\beta}{2} |\psi|^4 + \frac{1}{2m^*} \left| \left(-i\hbar\nabla - \frac{q^*}{c}\mathbf{A} \right) \psi \right|^2 + \frac{|\mathbf{\vec{\nabla} \times \vec{A}}|^2}{8\pi} $$

```
(*Com field B=curl A*)
eq1B =  1/(2 m) ((hbar/I Grad[\[Psi][x, y, z], {x, y, z}] - 
        q/c A \[Psi][x, y, z]).(-(hbar/I)
           Grad[\[Psi]c[x, y, z], {x, y, z}] - 
        q/c A \[Psi]c[x, y, z])) +
        1/(8 Pi) (Curl[A, {x, y, z}].Curl[A, {x, y, z}]) - 
 1/(4 Pi) (H.Curl[A, {x, y, z}]) // Expand
```

Doing the same procedure as before, $$\frac{\partial F}{\partial \psi}=0 \textrm{ and } \frac{\partial F}{\partial \psi^*}=0$$

```
eqx = VariationalD[eq1B, A1[x, y, z], {x, y, z}] // Expand;

eqy = VariationalD[eq1B, A2[x, y, z], {x, y, z}] // Expand;

eqz = VariationalD[eq1B, A3[x, y, z], {x, y, z}] // Expand;
{eqx,eqy,eqz}
```

```
fieldB = Curl[Curl[A, {x, y, z}], {x, y, z}];

currentJ= -(q^2/(m c) \[Psi][x, y, z] \[Psi]c[x, y, z] A -
   (q hbar)/(
    I 2 m) (\[Psi]c[x, y, 
        z] Grad[\[Psi][x, y, z], {x, y, z}] - \[Psi][x, y, 
        z] Grad[\[Psi]c[x, y, z], {x, y, z}]));

1/c (c/(4 Pi) fieldB - currentJ)
```

* Plot the Free energy for several temperatures assuming a uniform $\psi$ and zero magnetic field
* Plot the respective extreme solutions of Free energy as functions of the temperature (this will give the order parameter dependece on temperature)

The free energy equation is
$$F_{free}=\alpha|\psi|^2+\frac{\beta}{2}|\psi|^4$$
where $\psi$ is uniform ($\nabla|\psi|=0$) and no fields present.

Where we want to verify how the free energy equation changes with the $\alpha$. Let's use $\alpha_1>0,\alpha_2=0\textrm{ and }\alpha_3<0$, and $\beta>0$.

```
\[Alpha]1=0.1;
\[Alpha]2=0;
\[Alpha]3=-0.1;
\[Beta]=0.1;
```

Each function for each $\alpha$.

```
(* free energy *) (*Field zero*)
eq1 = \[Alpha]1(psi[x,y,z]^2) +
   1/2 \[Beta](psi[x,y,z]^4); (*Alpha positivo*)
eq2 = \[Alpha]2(psi[x,y,z]^2) +
   1/2 \[Beta](psi[x,y,z]^4); (*Alpha nulo*)
eq3 = \[Alpha]3(psi[x,y,z]^2) +
   1/2 \[Beta](psi[x,y,z]^4); (*Alpha negativo*)
```

```
Plot[{eq1,eq2,eq3},{psi[x,y,z],-2,2},AxesLabel -> {"|\[Psi]|","F"}, PlotLegends -> {"\[Alpha]>0","\[Alpha]=0","\[Alpha]<0"}]
```

The minimization condition is obtain by
$$\frac{\partial F}{\partial\psi^*}=0\Leftrightarrow 2\alpha\psi+2\beta|\psi|^2\psi=0\Leftrightarrow \psi(\alpha+\beta|\psi|^2)=0\Leftrightarrow \left\{\begin{array}{c} \psi=0 \\ |\psi|=\pm \sqrt{-\frac{\alpha}{\beta}} \end{array} \right.$$
Considering both solutions. And $\alpha$ can be given by $a_0(T-T_c)$.
After this substition we can determine how the minimization condition varies with the temperature.

```
Tc=100;
a0=0.1;
\[Alpha][T_]:=a0*(Tc-T); (*with a negative sign due to the sqrt(-alpha/beta)*)
\[Beta]=1;
```

```
solpsi0=0;
solpsi1[T_]:=Sqrt[\[Alpha][T]/\[Beta]];
```

```
Plot[{solpsi[T],-solpsi[T],solpsi0},{T,98,101},AxesLabel -> {"T","|\[Psi]|"},PlotLegends->{"|\[Psi]|>0","|\[Psi]|<0","|\[Psi]|=0"}]
```

__EXERCISE 3__

The Landau theory of phase transitions is based on the idea that the free energy can be expanded as a power series in the order parameter m. For a second order phase transition, the order parameter grows continuously from zero at the phase transition so the first few terms of the power series will dominate. If the free energy is expanded to sixth order in the order parameter, 

$$ F = F_n + \alpha |\psi|^2 + \frac{\beta}{2} |\psi|^4+ \frac{\gamma}{3} |\psi|^6  $$

the system will undergo a first order phase transition if $α_0>0$, $β<0$, and $γ>0$.

Plot the minimization extreme solutions of the free energy with respect to Temperature.

```
a0=+1;
Tc=100;
alpha[T_]:=a0*(T-Tc);
beta=-1;
gamma=+2;
```

```
alpha1=0.05;
alpha2=0;
alpha3=-0.1;
F[alpha_]:=alpha*x^2+beta*x^4/2+gamma*x^6/3;
```

```
Plot[{F[alpha1],F[alpha2],F[alpha3]},{x,-1,1},AxesLabel->{"|\[Psi]|","F"},PlotLegends->{"T>Tc","T=Tc","T<Tc"}]
```

Let's minimize the equation by deriving with respect to $\psi^*$, we get,$$\frac{dF}{d\psi^*}=\psi(\alpha+\beta|\psi|^2+\gamma|\psi|^4)=0 \Leftrightarrow |\psi|=\pm\sqrt{\frac{-\beta\pm\sqrt{\beta^2-4\alpha\gamma}}{2\gamma}} \textrm{ V } |\psi|=0,$$

with $\alpha(T)=\alpha_0(T-T_c)$, and $\alpha_0 > 0$, $\beta < 0$ and $\gamma > 0$. 

```
psi2plus[T_]:=Sqrt[(-beta+Sqrt[beta^2-4*alpha[T]*gamma])/(2*gamma)];
psi2minus[T_]:=-Sqrt[(-beta+Sqrt[beta^2-4*alpha[T]*gamma])/(2*gamma)];
psi2implus[T_]:=Sqrt[(-beta-Sqrt[beta^2-4*alpha[T]*gamma])/(2*gamma)];
psi2imminus[T_]:=-Sqrt[(-beta-Sqrt[beta^2-4*alpha[T]*gamma])/(2*gamma)];
psi0=0;
```

```
Plot[{psi2plus[T],psi2minus[T],psi2implus[T],psi2imminus[T],psi0},{T,98,101},AxesLabel -> {"T","|\[Psi]|"},PlotLegends -> {"+ +","- +","+ -","- -","|\[Psi]|=0"}]
```
