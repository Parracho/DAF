The free energy, with an uniform $\psi_i$, no fields, and a new coupling term, is described by,
$$F=\alpha(|\psi_1|^2+|\psi_2|^2)+\frac{\beta}{2}(|\psi_1|^4+|\psi_2|^4)+\nu(|\psi_1|^2|\psi_2|^2)=$$

To find the extreme points we take the derivative of the free energy with respect to $\psi_i$,$$\frac{dF}{d\psi_i}=\alpha\psi_i^*+\beta(\psi_i^*)^2\psi_i+\nu|\psi_j|^2\psi_i^*=0 \Leftrightarrow $$
$$\Leftrightarrow \psi_i^*(\alpha+\beta|\psi_i|^2+\nu|\psi_j|^2)=0\Leftrightarrow \psi_i^*=0\vee \alpha+\beta|\psi_i|^2+\nu|\psi_j|^2=0 $$
$$\Leftrightarrow \psi_i=0 \vee |\psi_i|^2=-\frac{1}{\beta}(\alpha+\nu|\psi_j|^2)$$
with $j\neq i$.
$$\left\{\begin{array}{c} |\psi_1|^2=-\frac{1}{\beta}(\alpha+\nu|\psi_2|^2) \\ |\psi_2|^2=-\frac{1}{\beta}(\alpha+\nu|\psi_1|^2)\end{array}\right. \Leftrightarrow |\psi_1|^2=|\psi_2|^2=\frac{-\alpha}{\nu+\beta}$$

```
alpha0 = 1;
beta = -10;
gama = 1;
v = 100;
Tc1 = 30;
Tc2= 15;
alpha1[T_] := alpha0 * (T-Tc1);
alpha2[T_] := alpha0 * (T-Tc2);
```

```
a = ((gama/v)*beta)^2;
b[T_] := 2*(gama/v)^2*alpha1[T]*beta - (gama/v)*beta^2 + gama*v;
c[T_] := (gama^2/v)*alpha1[T]^2 + gama*alpha2[T] - (gama/v)*beta*alpha1[T];


Psi1pp[T_] := +Sqrt[-b[T] + Sqrt[b[T]^2-4*a*c[T]]/(2*a)];
Psi1pm[T_] := +Sqrt[-b[T] - Sqrt[b[T]^2-4*a*c[T]]/(2*a)];
Psi1mp[T_] := -Sqrt[-b[T] + Sqrt[b[T]^2-4*a*c[T]]/(2*a)];
Psi1mm[T_] := -Sqrt[-b[T] - Sqrt[b[T]^2-4*a*c[T]]/(2*a)];



Psi2ppp[T_] :=  Sqrt[(-1/v) * (alpha1[T]+beta*Psi1pp[T]^2)];
Psi2ppm[T_] :=  Sqrt[(-1/v) * (alpha1[T]+beta*Psi1pm[T]^2)];
Psi2pmp[T_] :=  Sqrt[(-1/v) * (alpha1[T]+beta*Psi1mp[T]^2)];
Psi2pmm[T_] :=  Sqrt[(-1/v) * (alpha1[T]+beta*Psi1mm[T]^2)];
Psi2mpp[T_] :=  -Sqrt[(-1/v) * (alpha1[T]+beta*Psi1pp[T]^2)];
Psi2mpm[T_] :=  -Sqrt[(-1/v) * (alpha1[T]+beta*Psi1pm[T]^2)];
Psi2mmp[T_] :=  -Sqrt[(-1/v) * (alpha1[T]+beta*Psi1mp[T]^2)];
Psi2mmm[T_] :=  -Sqrt[(-1/v) * (alpha1[T]+beta*Psi1mm[T]^2)];
```
